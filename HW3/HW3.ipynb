```python
from __future__ import print_function
__author__= 'fbb'
import os
import numpy as np
import pylab as pl
%pylab inline
from scipy.optimize import curve_fit, minimize
```

```python
distributions = ['Poisson', 'Gaussian', 'ChiSquared', 'Cauchy', 'Logistic', 'Binomial']
```

```python
sample_size = np.arange(11,2000,20)
```

```python
md={}
for distb in distributions:
    md[distb]={}

for n in sample_size:
    md['Poisson'][n] = np.random.poisson(100, size=n)
    md['Cauchy'][n] = np.random.standard_cauchy(size=n)+100
    md['Gaussian'][n] = np.random.randn(n)+100    
    md['ChiSquared'][n] = np.random.chisquare(100,n)
    md['Binomial'][n] = np.random.binomial(200,0.5,n)
    md['Logistic'][n] = np.random.logistic(100,size=n)
```

```python
means={}
final_means=[]
for x in distributions:
    means[x] = {}
```

```python
# Create the figures
fig, ax = plt.subplots(figsize=(15,15), nrows=3, ncols=2)
for ix,distb in enumerate(distributions):
    
    for n in md[distb]:
#         print(ix,ix/2, ix%2)
#         print(n,type(n))
#         print(means[distb])

        means[distb][n] = md[distb][n].mean()
        ax[ix/2, ix%2].plot(n, means[distb][n], '.', c='tomato')
        ax[ix/2, ix%2].set_xlabel('Sample Size')
        ax[ix/2, ix%2].set_ylabel('Mean')
        ax[ix/2, ix%2].set_title(distb)
       
```

Given small samples, the distributions of means tend to vary more, i.e. they are more spread out. As sample sizes increase, the means appear to converge on the actual mean. This is what I would expect.
